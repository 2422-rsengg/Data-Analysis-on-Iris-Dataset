```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
X=dt[['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']]
y=dt['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
lm.coef_
```

```python
predictions=lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(dt.drop('target',axis=1))
```

```python
centers = kmeans.cluster_centers_
centers
```

```python
dt['klabels'] = kmeans.labels_
dt.head()
```

```python
f, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2,
                             sharey = True, figsize = (10,6))

# For fitted with kmeans 
ax1.set_title('K Means (K = 3)')
ax1.scatter(x = dt['sepal length (cm)'], y = dt['sepal width (cm)'], 
            c = dt['klabels'], cmap='rainbow')

ax1.scatter(x=centers[:, 0], y=centers[:, 1],
         c='black',s=300, alpha=0.5);

# For original data 
ax2.set_title("Original")
ax2.scatter(x = dt['sepal length (cm)'], y = dt['petal width (cm)'], 
            c = dt['target'], cmap='rainbow')
```

```python
# sum_square will be key,value pair for the elbow plot!
sum_square = {}
# Let's test for K from 1 to 10, we can use range() function here! remember?
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k).fit(dt.drop('target',axis=1))
    # .inertia: Computing Sum of Squared Distances of samples to their closest cluster center.
    sum_square[k] = kmeans.inertia_ 
```

```python
plt.plot(list(sum_square.keys()), list(sum_square.values()),
         linestyle ='-', marker = 'H', color = 'g',
         markersize = 8,markerfacecolor = 'b')
```
